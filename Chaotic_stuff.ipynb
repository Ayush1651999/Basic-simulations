lets take T = 1, so w = 2*pi

```python
import numpy as np
import matplotlib.pyplot as plt

pi = np.pi
w = 2*pi

a = np.linspace(0,2,301)
x_coor = np.zeros(201)
y_coor = np.zeros(201)

for i in range(201):
    x_coor[i] = np.sin(w*a[i]) + 4*np.sin(w*a[i]+pi/6) + 9*np.sin(w*a[i]+pi/2)
    y_coor[i] = np.sin(w*a[i] + pi/3) + 4*np.sin(w*a[i] + pi/3 + pi/6) + 15*np.sin(w*a[i] + pi/3 + pi/2)
    
plt.plot(x_coor,y_coor)
plt.show()
```
